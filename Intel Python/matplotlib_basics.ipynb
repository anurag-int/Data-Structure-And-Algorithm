```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,20)

y = x ** 2
#simple plot
plt.plot(x,y)
```

```python
#labeline the axis
plt.title("relationship between x and y")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y)
```

```python
#ofrmatting line type 

x = np.arange(0,20)
y = x ** 2
plt.title("Cubic relationship between x and y")
plt.plot(x,y,"r")
plt.plot(x,y,">")
```

```python
#adding legends
x = np.arange(0,10)
y = x ^ 2
z = x ^ 3
t = x ^ 4
plt.annotate('Second Entry',xy=[2,1])
plt.annotate('Third Entry',xy=[4,6])
plt.plot(x,z)
plt.plot(x,t)
plt.legend(['Race1', 'Race2', 'Race3'], loc=4)
```

```python
import numpy as np
from matplotlib import pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.plot(t1, f(t1), 'o', t2, f(t2), "r")
```

```python
# plotting pie graph

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True,startangle=0)
ax1.axis('equal')
plt.show()
```

```python
#bar graph
def plot_bar_x():
    index = np.arange(len(label))
    plt.bar(index, no_movies)
    plt.xlabel('Genre', fontsize=20)
    plt.ylabel('No of label', fontsize = 20)
    plt.xticks(index, label, fontsize=10, rotation=90)
    plt.title('Market share for Each genre 1995-2017', fontsize=25)
    plt.show()
label = ['Adventure', 'Action', 'Drama', 'Comedy', 'Thriller/Suspense', 'Horror' ,'Romantic']
no_movies = [941,854,3504,643,456,543,875]
plot_bar_x()
   
```

```python
x = np.linspace(0,5,11)
y = x ** 2
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,"g*-")
plt.xlabel("X Axis title here")
plt.ylabel("Y axis title here")
plt.title("String tiitle here")
```

```python
# another way of using subplots
fig, ax = plt.subplots()
ax.plot(x,x**2, "b*-")
ax.plot(x,x**3, "g--")

ax.set_xlabel("X axis")
ax.set_ylabel("y axis")
ax.set_title("plot title")
```

```python
#histogram with text

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x,50,density=1,facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('probability')

plt.title('Histrogram of IQ')
plt.text(60,0.025, '$ \mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
#plot different varities of graphs as subplots
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
n=np.array([0,1,2,3,4,5])
xx=np.linspace(-0.5,1.,100)
fig,axes=plt.subplots(1,4,figsize=(12,3))
axes[0].scatter(xx,xx+0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n,n**2,lw=2)
axes[1].set_title("step")
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)
axes[2].set_title("bar")
axes[3].fill_between(xx,xx*2,xx*3,color="green",alpha=0.9)
axes[3].set_title("fill_between")
```
